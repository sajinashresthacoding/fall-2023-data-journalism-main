## Single page Tabular Scrape

- <a href="https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action">On this site</a>, scrape all the doctors info on page 292.
- Export the content into a CSV file called ```page_292.csv```.

````python
## create cells as needed
import pandas as pd
import requests
````

````python
url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action?d-49653-p=292"
````

````python
response = requests.get(url)
response.status_code
````

````python
pd.read_html(response.text)[0]
````

````python
len(pd.read_html(response.text)[0])
````

````python
df_singlepage = pd.read_html(response.text)[0]
df_singlepage
````

````python
df_singlepage.to_csv("page_292.csv", encoding = "UTF-8", index = False)
````

## Multipage Tabular Scrape

- <a href="https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action">On this site</a>, scrape all doctors whose last names begin with "P".
- Export the content into a CSV file called ```md_P.csv```.


````python
## create cells as needed
base_url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AlphabetSearchAction.action?alpbhabetSearch=P*&d-49653-p="
````

````python
urls_lc = [(f"{base_url}{url_number}")for url_number in range (1,24)]

urls_lc
````

````python
from random import randrange
import time
````

````python
busted_links = []
df_all = []
total_links = len(urls_lc)
counter = 1 

for url in urls_lc:
    print(f"scraping {counter} of {total_links}")
    counter += 1
    response = requests.get(url)
    try:
        df = pd.read_html(response.text)[0]
    except:
        print(f"oh no! {url} returned {response.status_code}")
        busted_links.append(url)
    else:
        df_all.append(df)
        print(df)
    snoozer = randrange(30,45)
    print(f"snoozing for {snoozer} seconds before next scrape")
    time.sleep(snoozer)

print("done scraping all provided link")
````
