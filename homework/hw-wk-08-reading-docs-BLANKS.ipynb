````python
care = []

for text_file in all_txts:
    with open(text_file, "r") as my_text:
        all_text = my_text.readlines()
        year = all_text[0].replace("Year: " , "").replace("\n", "")
        cognition_related = all_text[2]
        if "Dementia" or "Alzheimer's" in cognition_related:
            cognition_related = "True"            
        else:
            cognition_related = "False"
        medical_conditions = all_text[2]
        if "determined to suffer from Alzheimer's" in medical_conditions:
            medical_conditions = "Alzheimer's"
        elif "determined to suffer from Dementia" in medical_conditions:
            medical_conditions = "Dementia"
        else:
            medical_conditions = "Not Specified"
        care_hours = all_text[4]
        if "12" in care_hours:
            care_hours = "half-day"
        elif "24" in care_hours:
            care_hours = "full-day"
        else:
            care_hours = "Not specified"
        care_dict = {"year": year, "cognition_related": cognition_related, "medical_condition": medical_conditions,
                    "care_hours": care_hours}
        care.append(care_dict)
````

````python
care
````

````python
df_care = pd.DataFrame(care)
df_care
````

````python
df_care.to_csv("care_given.csv", encoding = "UTF-8", index = False)
````
